You can represent angles as numbers (floats or integers). By default, most mathematical functions in Python's `math` module that deal with angles (like `sin`, `cos`, `tan`) expect the angle to be in **radians**.

```python
import math

# Angle in radians
angle_radians = math.pi / 2
print(f"Angle in radians: {angle_radians}")

# Angle in degrees
angle_degrees = 30
print(f"Angle in degrees: {angle_degrees}")
```

```python
# Let's say you have two angles
angle1 = 45  # in degrees
angle2 = 45  # in degrees

# To state that angle1 is equal to angle2, you can simply use the assignment operator if you are setting a value:
angle3 = angle1
print(f"Angle 3 is equal to Angle 1: {angle3}")

# To check if two angles are equal, you use the equality operator (==):
are_angles_equal = angle1 == angle2
print(f"Are Angle 1 and Angle 2 equal? {are_angles_equal}")

# If they are not equal
angle4 = 60
are_angles_equal_2 = angle1 == angle4
print(f"Are Angle 1 and Angle 4 equal? {are_angles_equal_2}")
```

To convert between degrees and radians, you can use the `math.degrees()` and `math.radians()` functions.

```python
import math

# Convert radians to degrees
angle_radians = math.pi
angle_degrees = math.degrees(angle_radians)
print(f"{angle_radians} radians is equal to {angle_degrees} degrees")

# Convert degrees to radians
angle_degrees = 180
angle_radians = math.radians(angle_degrees)
print(f"{angle_degrees} degrees is equal to {angle_radians} radians")
```

When using trigonometric functions, remember that they typically work with radians.

```python
import math

# Calculate the sine of a 90-degree angle (which is pi/2 radians)
angle_degrees = 90
angle_radians = math.radians(angle_degrees)
sine_value = math.sin(angle_radians)
print(f"The sine of {angle_degrees} degrees ({angle_radians} radians) is: {sine_value}")
```
